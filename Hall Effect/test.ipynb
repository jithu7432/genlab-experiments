```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.dpi']=95
print("Setup Complete")
```

```python
tab1 = pd.read_csv('tab1.csv')
tab2 = pd.read_csv('tab2.csv')
tab3 = pd.read_csv('tab3.csv')
tab4 = pd.read_csv('tab4.csv')
tab5 = pd.read_csv('tab5.csv')
```

```python
x = tab1['x']
y = tab1['y']
plt.plot(x, y,'-s',c = 'r')
plt.xlabel('Current (A)')
plt.ylabel('Magnetic Field (KG) ')
plt.grid(linestyle = '--')
plt.title('Calibration Curve')
slope = np.polyfit(x,y,1)[0].__round__(4)
plt.text(2.5,1.45,'Slope = {}'.format(slope),bbox=dict(facecolor='blue', alpha=0.3))
plt.savefig('Figure 1',dpi=400)
```

```python
x = tab2['x']
y = tab2['y']
plt.plot(x, y,'-s',c = 'r')
plt.ylabel('Hall Voltage (mV)')
plt.xlabel('Magnetic Field (KG) ')
plt.grid(linestyle = '--')
plt.suptitle('HALL VOLTAGE VS. MAGNETIC FIELD')
plt.title('(probe current = 2mA)',fontsize =10)
slope = np.polyfit(x,y,1)[0].__round__(4)
plt.text(4,24,'Slope = {}'.format(slope),bbox=dict(facecolor='blue', alpha=0.3))
plt.savefig('Figure 2',dpi=400)
```
