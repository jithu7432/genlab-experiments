```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
from scipy.stats import linregress
import seaborn as sns
plt.rcParams['figure.dpi']=300
print("Setup Complete")
```

```python
blue = pd.read_csv('blue.csv')
red = pd.read_csv('red.csv')
yellow = pd.read_csv('yellow.csv')
green = pd.read_csv('green.csv')
```

```python
x1 = blue['x']
y1 = blue['y']
x2 = red['x']
y2 = red['y']
x3 = yellow['x']
y3 = yellow['y']
x4 = green['x']
y4 = green['y']
```

```python
plt.plot(x1,y1,'-o',color = 'blue')
plt.plot(x2,y2,'-o',color = 'red')
plt.plot(x3,y3,'-o',color = 'yellow')
plt.plot(x4,y4,'-o',color = 'green')
legend = ['Blue','Red','Yellow','Green']
plt.grid()
plt.legend(legend)
plt.savefig('graph',dpi = 1000)
```
