```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import find_peaks

pd.plotting.register_matplotlib_converters()

plt.rcParams['figure.dpi'] = 100
plt.rcParams['axes.grid'] = True
plt.style.use('seaborn')
```

```python
data = pd.read_csv('dataset.csv')
```

```python
current = data['Current']
voltage = data['Voltage']
peaky = find_peaks(current)[0]
```

```python
plt.subplot(211)

sns.scatterplot(x = voltage, y = current, marker = 'X', color = 'r')
plt.plot(voltage, current, '--')
plt.xlabel("")
plt.ylabel("Current (Nano-Ampere)")
plt.title("ELECTRON BEAM CURRENT // ACCELERATING VOLTAGE",fontsize = 12)

plt.subplot(212)

plt.plot(voltage,current,'g--')
plt.xticks(peaky)
plt.ylabel("Current (Nano-Ampere)")
plt.xlabel("Accelerating Voltage (V)")


plt.tight_layout()
plt.savefig("Figure_1",dpi = 300)
```
